```python
import pandas as pd
import numpy as np
```

```python
#CRIMES TABLE

districts = ['ABP Sants-Montjuïc', 'ABP Ciutat Vella', "ABP Eixample", 'ABP Les Corts',
             'ABP Gràcia', 'ABP Sarrià-Sant Gervasi', 'ABP Sant Martí', 'ABP Sant Andreu',
             'ABP Nou Barris', 'ABP Horta-Guinardó']

district_codes = [1, 2, 6, 7, 4, 8, 9, 10, 3, 5]

crime_rates = pd.DataFrame()

# 2018 - Crime Rate

crime_rates_2018 = pd.read_excel('../datasets/crime rates 2018.xls')

crime_rates_2018 = crime_rates_2018.drop(['Mes', 'Any', 'Nom mes', 'Regió Policial (RP)', 'Títol Codi Penal',
                                'Tipus de fet', 'Resolts', 'Detencions'], axis=1)

crime_rates_2018 = crime_rates_2018.rename(columns={'Àrea Bàsica Policial (ABP)':'District',
                                          'Coneguts':'Crimes in 2018'})

crime_rates_2018 = crime_rates_2018[crime_rates_2018['District'].isin(districts)]

crime_rates_2018['District'] = [value[4:] for value in crime_rates_2018['District']]

crime_rates_2018 = crime_rates_2018.groupby(by=['District'], as_index=False).agg({'Crimes in 2018':np.sum})

crime_rates_2018['District Code'] = district_codes

crime_rates = crime_rates_2018[['District Code', 'District', 'Crimes in 2018']]

# 2017 - Crime Rate

crime_rates_2017 = pd.read_excel('../datasets/crime rates 2017.xls')

crime_rates_2017 = crime_rates_2017.drop(['Mes', 'Any', 'Nom mes', 'Regió Policial (RP)', 'Títol Codi Penal',
                                'Tipus de fet', 'Resolts', 'Detencions'], axis=1)

crime_rates_2017 = crime_rates_2017.rename(columns={'Àrea Bàsica Policial (ABP)':'District',
                                          'Coneguts':'Crimes in 2017'})

crime_rates_2017 = crime_rates_2017[crime_rates_2017['District'].isin(districts)]

crime_rates_2017['District'] = [value[4:] for value in crime_rates_2017['District']]

crime_rates_2017 = crime_rates_2017.groupby(by=[
    'District'], as_index=False).agg({'Crimes in 2017':np.sum})

crime_rates_2017 = crime_rates_2017[['District', 'Crimes in 2017']]

crime_rates['Crimes in 2017'] = crime_rates_2017['Crimes in 2017']

crime_rates
```

```python
# Crime oct 2017
crime_rates_2017 = pd.read_excel('../datasets/crime rates 2017.xls')

crime_rates_2017 = crime_rates_2017.loc[crime_rates_2017[
    'Mes'] == 10, ['Àrea Bàsica Policial (ABP)', 'Coneguts', 'Mes']] 

crime_rates_2017 = crime_rates_2017.rename(columns={'Àrea Bàsica Policial (ABP)':'District',
                                          'Coneguts':'Crimes in 10/2017', 'Mes' : 'Month'})

crime_rates_2017 = crime_rates_2017[crime_rates_2017['District'].isin(districts)]

crime_rates_2017['District'] = [value[4:] for value in crime_rates_2017['District']]

crime_rates_2017 = crime_rates_2017.groupby(by=['Month', 'District'], as_index=False).agg({
    'Crimes in 10/2017':np.sum})

crime_rates_oct = crime_rates_2017[[ 'District', 'Month', 'Crimes in 10/2017']]

# crime nov 2017
crime_rates_092017 = pd.read_excel('../datasets/crime rates 2018.xls')

crime_rates_092017 = crime_rates_092017.loc[crime_rates_092017[
    'Mes'] == 9, ['Àrea Bàsica Policial (ABP)', 'Coneguts', 'Mes']] 

crime_rates_092017 = crime_rates_092017.rename(columns={'Àrea Bàsica Policial (ABP)':'District',
                                          'Coneguts':'Crimes in 09/2017', 'Mes' : 'Month'})

crime_rates_092017 = crime_rates_092017[crime_rates_092017['District'].isin(districts)]

crime_rates_092017['District'] = [value[4:] for value in crime_rates_092017['District']]

crime_rates_092017 = crime_rates_092017.groupby(by=['Month', 'District'], as_index=False).agg({
    'Crimes in 09/2017':np.sum})

crime_rates_092017 = crime_rates_092017[[ 'District', 'Month', 'Crimes in 09/2017']]
crime_rates_oct['Crimes in 09/2017'] = crime_rates_092017['Crimes in 09/2017']

crime_rates_riots = crime_rates_oct[[
    'District', 'Crimes in 09/2017', 'Crimes in 10/2017']].sort_values(
    by='Crimes in 10/2017', ascending=False)

crime_rates_riots
```

```python
# 2016 - Crime Rates

crime_rates_2016 = pd.read_excel('../datasets/crime rates 2016.xls')

crime_rates_2016 = crime_rates_2016.drop(['Mes', 'Any', 'Nom mes', 'Regió Policial (RP)', 'Títol Codi Penal',
                                'Tipus de fet', 'Resolts', 'Detencions'], axis=1)

crime_rates_2016 = crime_rates_2016.rename(columns={'Àrea Bàsica Policial (ABP)':'District',
                                          'Coneguts':'Crimes in 2016'})

crime_rates_2016 = crime_rates_2016[crime_rates_2016['District'].isin(districts)]

crime_rates_2016['District'] = [value[4:] for value in crime_rates_2016['District']]

crime_rates_2016 = crime_rates_2016.groupby(by=['District'], as_index=False).agg({'Crimes in 2016':np.sum})

crime_rates_2016['District Code'] = district_codes

crime_rates['Crimes in 2016'] = crime_rates_2016[['Crimes in 2016']]

#crime rates 2015

crime_rates_2015 = pd.read_excel('../datasets/crime rates 2015.xls')

crime_rates_2015 = crime_rates_2015.drop(['Mes', 'Any', 'Nom mes', 'Regió Policial (RP)', 'Títol Codi Penal',
                                'Tipus de fet', 'Resolts', 'Detencions'], axis=1)

crime_rates_2015 = crime_rates_2015.rename(columns={'Àrea Bàsica Policial (ABP)':'District',
                                          'Coneguts':'Crimes in 2015'})

crime_rates_2015 = crime_rates_2015[crime_rates_2015['District'].isin(districts)]

crime_rates_2015['District'] = [value[4:] for value in crime_rates_2015['District']]

crime_rates_2015 = crime_rates_2015.groupby(by=[
    'District'], as_index=False).agg({'Crimes in 2015':np.sum})

crime_rates['Crimes in 2015'] = crime_rates_2015[['Crimes in 2015']]

crime_rates
```

```python
#crime rates 2013

crime_rates_2013 = pd.read_excel('../datasets/crime rates 2013.xls')

crime_rates_2013 = crime_rates_2013.drop(['Mes', 'Any', 'Nom mes', 'Regió Policial (RP)', 'Títol Codi Penal',
                                'Tipus de fet', 'Resolts', 'Detencions'], axis=1)

crime_rates_2013 = crime_rates_2013.rename(columns={'Àrea Bàsica Policial (ABP)':'District',
                                          'Coneguts':'Crimes in 2013'})

crime_rates_2013 = crime_rates_2013[crime_rates_2013['District'].isin(districts)]

crime_rates_2013['District'] = [value[4:] for value in crime_rates_2013['District']]

crime_rates_2013 = crime_rates_2013.groupby(by=[
    'District'], as_index=False).agg({'Crimes in 2013':np.sum})

crime_rates['Crimes in 2013'] = crime_rates_2013['Crimes in 2013']

crime_rates
```

```python
# ordered and merged

crime_rates['Crime Increase (2013-2018)'] = round(
    (crime_rates['Crimes in 2018'] - crime_rates['Crimes in 2013'])/
     crime_rates['Crimes in 2013'] * 100, 2)

crime_rates_sorted = crime_rates.sort_values(
    'Crime Increase (2013-2018)', ascending=False)

#ordered by index to merge
crime_rates_to_merge = crime_rates.set_index('District Code').sort_values(by=['District Code'])

crime_rates_sorted
crime_rates[['District', 'Crimes in 2013', 'Crimes in 2018', 'Crime Increase (2013-2018)']].sort_values(
    by='Crime Increase (2013-2018)', ascending=False)
```

```python
# POPULATION TABLE

population_data = pd.read_csv('../datasets/population.csv')

less_columns_population = population_data.loc[population_data['Year'] == 2017, 
                                                    ['Year', 'District.Code', 'District.Name',
                                                     'Gender', 'Age', 'Number']] 

# filtered out the people who are too young to work or retiered 

working_age = ['15-19', '20-24', '25-29', '30-34', '35-39', '40-44',
               '45-49', '50-54', '55-59','60-64']

working_population = less_columns_population[less_columns_population['Age'].isin(working_age)] 

# summed the age, so the district gets the entire numer of 'working population'

agg_column_population_without_genders = working_population.groupby(by=[
    'Year','District.Name', 'District.Code'], as_index =False ).agg({'Number':np.sum})

population = agg_column_population_without_genders.drop('Year', axis=1).rename(
    columns={'Number':'Population in 2017'}).set_index('District.Code')

# Population 2016

population_2016 = population_data.loc[population_data[
    'Year'] == 2016, ['Year', 'District.Code', 'District.Name', 'Gender', 'Age', 'Number']]

working_population2016 = population_2016[population_2016['Age'].isin(working_age)] 

# summed the age, so the district gets the entire numer of 'working population'

agg_column_population_without_genders2016 = working_population2016.groupby(by=[
    'Year','District.Name', 'District.Code'], as_index =False ).agg({'Number':np.sum})

# summed the age, so the district gets the entire numer of 'working population'

agg_column_population_without_genders2016 = working_population2016.groupby(by=[
    'Year','District.Name', 'District.Code'], as_index =False ).agg({'Number':np.sum})

population_2016 = agg_column_population_without_genders2016.drop(
    'Year', axis=1).rename(columns={'Number':'Population in 2016'}).set_index('District.Code')

#Population 2013

population_2013 = population_data.loc[population_data[
    'Year'] == 2013, ['Year', 'District.Code', 'District.Name', 'Gender', 'Age', 'Number']]

working_population2013 = population_2013[population_2013['Age'].isin(working_age)] 

# summed the age, so the district gets the entire numer of 'working population'

agg_column_population_without_genders2013 = working_population2013.groupby(by=[
    'Year','District.Name', 'District.Code'], as_index =False ).agg({'Number':np.sum})

# summed the age, so the district gets the entire numer of 'working population'

agg_column_population_without_genders2013 = working_population2013.groupby(by=[
    'Year','District.Name', 'District.Code'], as_index =False ).agg({'Number':np.sum})

population_2013 = agg_column_population_without_genders2013.drop(
    'Year', axis=1).rename(columns={'Number':'Population in 2013'}).set_index('District.Code')

population['Population in 2013'] = population_2013['Population in 2013'] 
population = population.sort_values(by=['District.Code'])
```

```python
#POPULATION INCREASE

population['Population Increase'] = round(
    (population['Population in 2017'] - population[
        'Population in 2013'])/population['Population in 2013'] * 100, 2)

population[['District.Name', 'Population Increase']].sort_values(
    'Population Increase', ascending=False)
```

```python
# Density

ordered_district_size = [4.49, 7.46, 21.35, 6.08, 20.09, 4.19, 11.96, 8.04, 6.56, 10.8]
#for size in ordered_district_size
density = round(population['Population in 2017']/ordered_district_size/1000, 2)

average_crime = crime_rates['Crime Increase (2013-2018)'].mean()
print(average_crime)

#population x crime
pop_crime = crime_rates_to_merge[[
    'District Code','District', 'Crime Increase (2013-2018)']].set_index(['District Code'])
pop_crime['Density - m2'] = density
print(pop_crime)
pop_crime.sort_values(by=['Density - m2'], ascending=False)
```

```python
#UNEMPLOYMENT RATES
#filtered by demand_occupation

data = pd.read_csv('../datasets/unemployment.csv')

#2017

filtered = data.loc[data['Year'] == 2017, 
                    ['Year', 'Month', 'District Code', 'District Name', 'Gender',
                     'Demand_occupation', 'Number']]

filtered = filtered.loc[filtered['Demand_occupation'] == 'Registered unemployed', 
                    ['Year', 'Month', 'District Code', 'District Name', 'Gender', 'Demand_occupation', 'Number']]

#aggregated registered number of unemployed-female and male

aggregated = filtered.groupby(by= ['Year','District Name','District Code', 'Demand_occupation'], 
                              as_index =False ).agg({'Number':(np.sum)})

aggregated['Number'] = round(aggregated['Number']/12, 2)

unemployed = aggregated.rename(columns={"Number" : "Unemployed in 2017"})

unemployed = unemployed[['District Name','District Code','Unemployed in 2017']]

unemployed_final = unemployed.set_index('District Code')

unemployed_final = unemployed_final.drop(99, axis=0)

# 2016

filtered2016 = data.loc[data['Year'] == 2016, 
                    ['Year', 'Month', 'District Code', 'District Name', 'Gender',
                     'Demand_occupation', 'Number']]

filtered2016 = filtered2016.loc[filtered2016['Demand_occupation'] == 'Registered unemployed', 
                    ['Year', 'Month', 'District Code', 'District Name', 'Gender', 'Demand_occupation', 'Number']]

#aggregated registered number of unemployed-female and male

aggregated2016 = filtered2016.groupby(by= ['Year','District Name','District Code', 'Demand_occupation'], 
                              as_index =False ).agg({'Number':np.sum})

aggregated2016['Number'] = round(aggregated2016['Number']/12, 2)

unemployed2016 = aggregated2016.rename(columns={"Number" : "Unemployed in 2016"})

unemployed2016 = unemployed2016[['District Name','District Code','Unemployed in 2016']]

unemployed_final2016 = unemployed2016.set_index('District Code')

unemployed_final2016 = unemployed_final2016.drop(99, axis=0)

# 2013

filtered2013 = data.loc[data['Year'] == 2013, 
                    ['Year', 'Month', 'District Code', 'District Name', 'Gender',
                     'Demand_occupation', 'Number']]

filtered2013 = filtered2013.loc[filtered2013[
    'Demand_occupation'] == 'Registered unemployed', [
    'Year', 'Month', 'District Code', 'District Name', 'Gender', 'Demand_occupation', 'Number']]

#aggregated registered number of unemployed-female and male

aggregated2013 = filtered2013.groupby(by= [
    'Year','District Name','District Code', 'Demand_occupation'], as_index =False ).agg({
    'Number':np.sum})

aggregated2013['Number'] = round(aggregated2013['Number']/12, 2)

unemployed2013 = aggregated2013.rename(columns={"Number" : "Unemployed in 2013"})

unemployed2013 = unemployed2013[['District Name','District Code','Unemployed in 2013']]

unemployed_final2013 = unemployed2013.set_index('District Code')

unemployed_final2013 = unemployed_final2013.drop(99, axis=0)

unemployed_final['Unemployed in 2013'] = unemployed_final2013['Unemployed in 2013']

unemployed_final = unemployed_final[['District Name', 'Unemployed in 2013', 'Unemployed in 2017']]

#Unemployment decrease

#unemp_decrease = round((unemployed_final['Unemployed in 2017'] - unemployed_final[
#    'Unemployed in 2013'])/unemployed_final['Unemployed in 2017']*100, 2)

# Unemployment Rate

unemp_pop_2017 = round(100-(population['Population in 2017'] - unemployed_final[
    'Unemployed in 2017'])/population['Population in 2017']*100, 2)

unemp_pop_2013 = round(100-(population['Population in 2017'] - unemployed_final[
    'Unemployed in 2013'])/population['Population in 2017']*100, 2)

unemployed_final['Unemployment Decrease'] = unemp_pop_2017 - unemp_pop_2013
unemployed_final['Unemployment Rate - 2017'] = unemp_pop_2017
unemployed_final['Unemployment Rate - 2013'] = unemp_pop_2013

unemployed_to_merge = unemployed_final.sort_index

unemployed_final = unemployed_final.sort_values(by=['Unemployment Decrease'], ascending=False)

unemployed_final[['District Name', 'Unemployment Rate - 2013',
                  'Unemployment Rate - 2017', 'Unemployment Decrease']]
```

```python
#IMMIGRATION RATES

immigration_data = pd.read_csv('../datasets/immigrants_by_nationality.csv')

#2017

filtered_im = immigration_data.loc[immigration_data['Year'] == 2017, ['District Name', 'Number']]

aggregated_im = filtered_im.groupby(by= ['District Name'], 
                              as_index =False ).agg({'Number':(np.sum)})

immigrants = aggregated_im.rename(columns={"Number" : "Immigration in 2017"}).drop(5)

#2015

filtered_im15 = immigration_data.loc[immigration_data['Year'] == 2015, ['District Name', 'Number']]

aggregated_im15 = filtered_im15.groupby(by= ['District Name'], 
                              as_index =False ).agg({'Number':(np.sum)})

immigrants15= aggregated_im15.rename(columns={"Number" : "Immigration in 2015"}).drop(5)

immigrants['Immigration in 2015'] = immigrants15['Immigration in 2015']

im_increase = round((immigrants['Immigration in 2017'] - immigrants[
    'Immigration in 2015'])/immigrants['Immigration in 2017']*100, 2)

immigrants['Immigration Increase'] = im_increase

immigrants_increase = immigrants[['District Name', 'Immigration Increase']]
    
#.sort_values('Immigration in 2017', ascending=False)
average_im = immigrants_increase['Immigration Increase'].mean()
print(average_im)

immigrants_increase
```
